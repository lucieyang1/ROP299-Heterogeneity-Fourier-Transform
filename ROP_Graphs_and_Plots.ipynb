```python
import matplotlib.pyplot as plt
import numpy as np

data = [
    (2, 1, 53.75, 56.75),
    (2, 2, 3.75, 142.00),
    (2, 3, 0.25, 42.00),
    (2, 4, 190.25, 15.25),
    (4, 1, 66.75, 53.50),
    (4, 2, 29, 25.75),
    (4, 3, 97, 2.75),
    (4, 4, 55.25, 174.00),
    (6, 1, 109.5, 7.00),
    (6, 2, 15.25, 173.00),
    (6, 3, 23.25, 25.75),
    (6, 4, 100, 50.25),
    (8, 1, 53.25, 120.25),
    (8, 2, 54.75, 1.75),
    (8, 3, 111.5, 17.50),
    (8, 4, 28.5, 116.50),
    (10, 1, 103.75, 131.00),
    (10, 2, 44, 76.50),
    (10, 3, 41, 11.50),
    (10, 4, 59.25, 37.00)
]

body_parts = [2, 4, 6, 8, 10]
clusters = [1, 2, 3, 4]
labels = [f'{bp}' for bp in body_parts]

plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.figsize'] = [8, 4]

num_clusters = len(clusters)
bar_width = 0.2
bar_offsets = np.arange(len(body_parts)) * (num_clusters + 1) * bar_width

fig, ax = plt.subplots()
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']

legend_labels = ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4']

for i, cluster in enumerate(clusters):
    size_data = [item[2] for item in data if item[1] == cluster]
    central_data = [item[3] for item in data if item[1] == cluster]
    
    size_rects = ax.bar(bar_offsets + i * bar_width, size_data, bar_width, label=f'Cluster {cluster} - Size', color=colors[i])
    central_rects = ax.bar(bar_offsets + i * bar_width, central_data, bar_width, bottom=size_data, color=colors[i], alpha=0.5)
    
#     legend_labels.append(f'Cluster {cluster} - Size')
#     legend_labels.append(f'Cluster {cluster} - Central')

# Create custom legend with correct colors
custom_legend = [plt.Line2D([0], [0], color=color, lw=4) for color in colors]
ax.legend(custom_legend, legend_labels, loc='upper center', ncol=4, bbox_to_anchor=(0.5, -0.15))

ax.set_ylabel('Size')
ax.set_xlabel('Number of Body Parts')
ax.set_title('CNN-extracted Features\' Cluster Sizes and Composition')
ax.set_xticks(bar_offsets + (num_clusters - 1) * bar_width / 2)
ax.set_xticklabels(labels)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Body Parts': [2, 4, 6, 8, 10],
    'Coefficient of Variation (CV)': [3.133, 17.348, 12.792, 7.741, 14.558]
}

# Create a scatter plot
fig, ax = plt.subplots()

# Plot scatter plot
ax.scatter(data['Body Parts'], data['Coefficient of Variation (CV)'], color='blue', label='Data')

# Fit a linear trendline
slope, intercept = np.polyfit(data['Body Parts'], data['Coefficient of Variation (CV)'], 1)
ax.plot(data['Body Parts'], slope * np.array(data['Body Parts']) + intercept, color='red', label='Trendline')

# Set labels and title
ax.set_xlabel('Number of Body Parts')
ax.set_ylabel('Coefficient of Variation')
ax.set_title('Coefficient of Variation as the Number of Body Parts Increase')

# Add legend
# ax.legend()

plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

# Data
body_parts = np.array([2, 4, 6, 8, 10])
cv = np.array([3.133, 17.348, 12.792, 7.741, 14.558])

# Calculate the linear regression
slope, intercept, r_value, p_value, std_err = linregress(body_parts, cv)

# Calculate R-squared
r_squared = r_value ** 2

print("R-squared value:", r_squared)
```
